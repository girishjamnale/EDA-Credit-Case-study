# Univariate Analysis


The data consists of people who have difficulties with payment ,which is denoted by a variable, TARGET , as 1 and those who make the payment on time,which is denoted by a variable, TARGET , as 0. Now we are going to check if there is huge difference between the number of applicants in both the cases

```python
#For this we would be considering the following:
target_0_perc=100*(len(final_app_data.loc[final_app_data['TARGET']==0])/len(final_app_data))
target_1_perc=100*(len(final_app_data.loc[final_app_data['TARGET']==1])/len(final_app_data))
sizes=[target_0_perc,target_1_perc]
target_labels=['TARGET=0','TARGET=1']
plt.pie(sizes, labels=target_labels,
autopct='%1.1f%%', startangle=90)
plt.show()
```

From the above pie chart we can observe that there is a huge imbalance in the data since only 8.7% of the data we have is of TARGET=1 . Hence the analysis will be hugely in favour of TARGET=0. Inorder to analyse this imbalnce in data, we use a mix of univariate and bivariate analysis on our data .

Now, we are going to divide the data into 2 data frames; one for TARGET=0 and the other for TARGET=1

```python

target_0=final_app_data.loc[final_app_data['TARGET']==0]
target_1=final_app_data.loc[final_app_data['TARGET']==1]
```

```python

target_1.info()
```

Since the imbalance percentage is cleary huge, we are going to check the data of teh selected variables for the entire application data set, i.e. including TARGET=0 and TARGET=1 w.r.t to the data of TARGET=1 (Since we are interested in the defaulters more than the ones who are able to repay the loan on time) Due to the huge imbalance in data we will be considering analysis for the entire application set and TARGET=1

```python
#Here , I would be first considering the univariate analysis for catgeorical variables for target==1
#By using the frequency of a category we plot the data for the columns which are of categorical types.
#Since we would be considering 10 categories, we created a common function for plotting the insights.
def uni_categorical_plot(col1,col2):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    sns.countplot(col1,order=col1.value_counts(ascending=False).index)
    plt.xticks(rotation=90)
    plt.title("Target=0")
    plt.subplot(1,2,2)
    sns.countplot(col2,order=col2.value_counts(ascending=False).index)
    plt.xticks(rotation=90)
    plt.title("Target=1")
    plt.show()
```

```python
uni_categorical_plot(target_0['NAME_CONTRACT_TYPE'],target_1['NAME_CONTRACT_TYPE'])
uni_categorical_plot(target_0['CODE_GENDER'],target_1['CODE_GENDER'])
```
